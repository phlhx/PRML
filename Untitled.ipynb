```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mn

NUM_C_1 = 1000
NUM_C_2 = 1000

##### fitting parameters #####
def fit_parameters(C_array, is_share = True):
    class_num = len(C_array)
    num = np.array([len(c) for c in C_array])
    u_array = num / num.sum()
    mu_array  = np.array([c.mean(axis = 0) for c in C_array])    
    sigma_array  = np.array([(C_array[i] - mu_array[i]).T.dot(C_array[i] - mu_array[i]) / num[i] for i in range(class_num)])
    
    if is_share:
        shared_sigma = 0
        for i in range(class_num):
            shared_sigma += (C_array[i] - mu_array[i]).T.dot(C_array[i] - mu_array[i]) 
            sigma = [shared_sigma / num.sum()] * class_num
        return u_array, mu_array,np.array(sigma)
    
    return u_array, mu_array, sigma_array
        
                                    
def plot_classes(C_array, u_array, mu_array, sigma_array, ax):
    class_num = len(C_array)
    X = np.arange(-13, 10, 0.1)
    Y = np.arange(-13, 10, 0.1)
    X, Y = np.meshgrid (X, Y)
    X_tmp = np.concatenate((np.reshape(X, X.shape + (1,)), np.reshape(Y, Y.shape + (1, ))), axis = 2)
    mn_list = [mn(mu_array[i], sigma_array[i]) for i in range(class_num)]
    Z_array = np.array([u_array[i] * mn_list[i].pdf(X_tmp) for i in range(class_num)])
    Z = np.argmax(Z_array, axis = 0)
    ax.contourf(X, Y, Z, cmap = "jet", alpha = 0.3)
    for i in range(class_num):
        ax.scatter(C_array[i].T[0], C_array[i].T[1])
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    
```

```python
fig, axes = plt.subplots(1, 3, figsize = (20, 6))
C_1 = np.random.multivariate_normal([-3,2.5], [[3,1],[1,1]] , NUM_C_1)
C_2 = np.random.multivariate_normal([-1,-2], [[3,1],[1,1]] , NUM_C_2)
C_3 = np.random.multivariate_normal([3,2], [[3,1],[1,1]] , NUM_C_2)
C_array = np.array([C_1, C_2, C_3])
u_array, mu_array, sigma_array = fit_parameters(C_array)
plot_classes(C_array, u_array, mu_array, sigma_array, axes[0])
axes[0].set_title("Shared Sigma with \n linear decision boundary (500 vs 500)", fontsize = 15)

C_2 = np.random.multivariate_normal([-1,-2], [[1, 0.2],[0, 1]] , NUM_C_2)
C_array = np.array([C_1, C_2, C_3])
u_array, mu_array, sigma_array = fit_parameters(C_array, False)
plot_classes(C_array, u_array, mu_array, sigma_array, axes[1])
axes[1].set_title("Individual Sigmas with \n nonlinear decision boundary (500 vs 500)", fontsize = 15)

C_1 = np.random.multivariate_normal([-3,2.5], [[3,1],[1,1]] , NUM_C_1 * 100)
C_array = np.array([C_1, C_2, C_3])
u_array, mu_array, sigma_array = fit_parameters(C_array, False)
plot_classes(C_array, u_array, mu_array, sigma_array, axes[2])
axes[2].set_title("Individual Sigmas with \n nonlinear decision boundary (50000 vs 500)", fontsize = 15)

```
