# EM Algorithm

"EM" in EM algorithm stand for "expectation" and "maximization", respectively. It is usually applied in the case of hidden features. For instance, usually we need to maximize likelihood function $P(X|\boldsymbol{\theta}) = \prod_{n=1}^NP(\boldsymbol{x}_n|\boldsymbol{\theta})$ ($\boldsymbol{\theta}$ is the parameters). Sometimes there will be more convenient to introduce some hidden features $Z$, the the likelihood function will be

$$
\prod_{n=1}^N\sum_{\boldsymbol{z}_n}P(\boldsymbol{x}_n, \boldsymbol{z}_n |\boldsymbol{\theta}) 
$$

We have made use of $P(\boldsymbol{x}_n|\boldsymbol{\theta})=\sum_{\boldsymbol{z}_n}P(\boldsymbol{x}_n, \boldsymbol{z}_n |\boldsymbol{\theta})$. The sum is annoying since it will be kept within logarithm as followed:

$$
\ln \prod_{n=1}^N\sum_{\boldsymbol{z}_n}P(\boldsymbol{x}_n, \boldsymbol{z}_n |\boldsymbol{\theta}) 
= \sum_{n=1}^N \ln \sum_{\boldsymbol{z}_n}P(\boldsymbol{x}_n, \boldsymbol{z}_n |\boldsymbol{\theta}) 
$$

Or in another view, sometimes it will be easier to do optimization on $P(X,Z|\boldsymbol{\theta})$ then on $P(X|\boldsymbol{\theta})$, but the problem is that we **DO NOT** know $Z$ in our training data and we wish we could have it. EM algorithm is designed to tackle such problem.

If we initialize an distribution with respect to $Z$ as $q(Z)$, and $\sum_{Z}q(Z) = 1$, then 

$$
\begin{align}
\ln P(X|\boldsymbol{\theta}) &= (\sum_{Z}q(Z)) \ln P(X|\boldsymbol{\theta}) \\
&=\sum_{Z}q(Z) \ln \frac{P(X, Z|\boldsymbol{\theta})}{P(Z|X, \boldsymbol{\theta})} \\
&=\sum_{Z}q(Z) \ln \frac{P(X, Z|\boldsymbol{\theta})}{q(Z)} - \sum_{Z}q(Z) \ln \frac{P(Z|X, \boldsymbol{\theta})}{q(Z)} \\
&=\mathscr{L}(q, \boldsymbol{\theta}) + KL(q||p)
\end{align}
$$

It should be noted that $KL(q||p) \geq 0$. If $KL(q||p) = 0$, then $\mathscr{L}(q, \boldsymbol{\theta}) = \ln P(X|\boldsymbol{\theta})$. In this way, we could maximize $\ln P(X|\boldsymbol{\theta})$ with respect to $\boldsymbol{\theta}$ by maximizing $\mathscr{L}(q, \boldsymbol{\theta}) $. Maximizing $\mathscr{L}(q, \boldsymbol{\theta})$ should be easier since it contains $P(X,Z|\boldsymbol{\theta})$. 

And to maximize $\sum_{Z}q(Z) \ln \frac{P(X, Z|\boldsymbol{\theta})}{q(Z)}$ with respect to $\boldsymbol{\theta}$ is equivalent to maximize $\sum_{Z}q(Z) \ln P(X, Z|\boldsymbol{\theta})$, because $\sum_{Z}q(Z)\ln q(Z)$ is a constant with respect to $\boldsymbol{\theta}$. And it should be noticed that $\sum_{Z}q(Z) \ln P(X, Z|\boldsymbol{\theta})$ is an expectation value of $\ln P(X, Z|\boldsymbol{\theta})$, because $q(Z)$ is a distribution. 

$$
KL(q||p) = 0 \Rightarrow q(Z) = P(Z|X, \boldsymbol{\theta})
$$

Usually to obtain $P(Z|X, \boldsymbol{\theta})$, it will make use of sum and product rules: 

$$
P(Z|X, \boldsymbol{\theta}) 
= \frac{P(X, Z| \boldsymbol{\theta})}{\sum_{Z}P(X, Z| \boldsymbol{\theta})}
$$

**Note that $\ln P(X|\boldsymbol{\theta})$ will not change by changing q(Z) ($\ln P(X|\boldsymbol{\theta})$ is the function of only $\boldsymbol{\theta}$)**.



-----
**EM Algorithm**
1. Initialize $\boldsymbol{\theta}^{old}$
2. **E-step**: Set $q = P(Z|X, \boldsymbol{\theta}^{old})$
3. **M-step**: Maximize $\mathbb{E}_Z(\boldsymbol{\theta}) = \sum_{Z}q(Z) \ln P(X, Z|\boldsymbol{\theta})$ with respect to $\boldsymbol{\theta}$, $\boldsymbol{\theta}^{new} = argmax \mathbb{E}_Z(\boldsymbol{\theta})$
4. Repeat 2, 3 steps until $\mathbb{E}_Z(\boldsymbol{\theta}) $ converges
-----

**EM Algorithm can only guarantee the local optimization instead of the global one,** which means the initialization is an important step.

The classical applications of EM algorithm are **Guassian mixture model** and **K-means**. K-means can be regarded as a special case of Guassian mixture model which will be discussed as followed

## Guassian mixture model
### Algebra
The name of Guassian misture model tells itself. It is a model of several guassian distribution mixing. Supposed we already know which Guassian distribution $\boldsymbol{x}_n$ belongs to and mark it as $\boldsymbol{z}_n$. For instance, if there are three Guassian distribution and $\boldsymbol{x}_n$ belongs to the second one, then $\boldsymbol{z}_n = (0, 1, 0)$.

Now let us apply EM algorithm:

**E-step**

We need to get $q(Z) = P(Z|X, \boldsymbol{\theta})$. In the Guassian mixture model, $\boldsymbol{\theta}$ includes $\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K$

$$
P(Z|X, \boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K) 
\propto P(X, Z|\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K)
=\prod_{n=1}^N\prod_{j=1}^K(\pi_j\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j))^{z_{nj}}
$$

**M-step**

First of all, we need to get $\mathbb{E}_Z(\boldsymbol{\theta})$:

$$
\begin{align}
\mathbb{E}_Z(\boldsymbol{\theta}) &= \mathbb{E}_Z(\ln P(X, Z|\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K)) \\
&= \mathbb{E}_Z(\sum_{n=1}^N\sum_{j=1}^Kz_{nj}(\ln \pi_j + \ln \mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j))) \\
&= \sum_{n=1}^N\sum_{j=1}^K\mathbb{E}_Z(z_{nj})(\ln \pi_j + \ln \mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j)) \\
\end{align}
$$

Now **the key** is $\mathbb{E}_Z(z_{nj})$:

$$
\begin{align}
\mathbb{E}_Z(z_{nj}) 
&= \frac{\sum_{z_{nj}}z_{nj}\prod_{m=1}^N\prod_{i=1}^K(\pi_i\mathscr{N}(\boldsymbol{x}_m|\boldsymbol{\mu}_i, \Sigma_i))^{z_{mi}}}
{\sum_{z_{nj}}\prod_{m=1}^N\prod_{i=1}^K(\pi_i\mathscr{N}(\boldsymbol{x}_m|\boldsymbol{\mu}_i, \Sigma_i))^{z_{mi}}} \\
&= \frac{\sum_{z_{nj}}z_{nj}(\pi_j\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j))^{z_{nj}}}
{\sum_{z_{nj}}(\pi_j\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j))^{z_{nj}}} \\
& = \frac{\pi_j\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j)}{\sum_{k}^K\pi_k\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_k, \Sigma_k)} \\
& \equiv \gamma(z_{nj})
\end{align}
$$

Therefore we have:

$$
\mathbb{E}_Z(\boldsymbol{\theta}) = \mathbb{E}_Z(\ln P(X, Z|\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K)) =\sum_{n=1}^N\sum_{j=1}^K\gamma(z_{nj})(\ln \pi_j + \ln \mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \Sigma_j))
$$

Just regard **$\gamma(z_{nj})$ as a constant** and set the derivative of $\mathbb{E}_Z(\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K)$ with respect to $\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K$ as zeros, then we could have:

$$\boldsymbol{\mu}_k = \frac{1}{N_k}\sum_{n=1}^N\gamma(z_{nk})\boldsymbol{x}_n $$
$$\Sigma_k = \frac{1}{N_k}\sum_{n=1}^N\gamma(z_{nk})(\boldsymbol{x}_n - \boldsymbol{\mu}_k)(\boldsymbol{x}_n - \boldsymbol{\mu}_k)^T$$
$$\boldsymbol{\pi}_k = \frac{N_k}{N} $$

where $N_k = \sum_{n=1}^N\gamma(z_{nk})$

To summarize the process:
1. Initialize $\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K$
2. Assign $\gamma(z_{nk})$
3. Assign $\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K$ by $\gamma(z_{nk})$
4. Repeat 2, 3 until $\mathbb{E}_Z(\ln P(X, Z|\boldsymbol{\pi}, \boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K, \Sigma_1, ..., \Sigma_K))$ converges

The number of clusters K is a hyperparameter which cannot be fitted. To determine the best K, a classical method is to plot loss function value versus K. The curve will have a "elbow" shape and the best K is right at the turning point.

Having trained the model, a new observation will be assigned to the k class if $\pi_k\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_k, \Sigma_k)$ has the largest value

### Simulation

In the simulation, by different parameters initialization, the final results are different due to the fact that EM algorithm only search for the local optimization.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mn
from matplotlib.ticker import MaxNLocator

def initialize_mu(X, k):
    dim = X.shape[-1]
    bounds = np.concatenate((X.min(axis = 0).reshape(-1, 1), X.max(axis = 0).reshape(-1, 1)), axis = -1)
    mu = []
    for d in range(dim):
        mu.append(np.random.uniform(bounds[d][0], bounds[d][1], k))
    mu_array = np.array(mu).T
    return mu_array

def EM(X, k, iteration = 100, is_random = True, preset_centers = None):
    ## Initialize pi, mu and Sigma; pis are set to be equally distributed; Sigmas are all set to be indentical matrix; 
    dim = X.shape[-1]
    N = len(X)
    if is_random:
        mu_array = initialize_mu(X, k)      
    else:
        mu_array = np.array(preset_centers)
    
    mu_init = mu_array.copy()
        
    sigma_array = np.array([np.eye(dim) for _ in range(k)])
    pi_array = np.ones(k) * (1/k)
    
    pi_init, mu_init, sigma_init = pi_array.copy(), mu_array.copy(), sigma_array.copy()
    
    for _ in range(iteration):
        ## assign gamma ##
        mn_tmp = [mn(mu_array[i], sigma_array[i]) for i in range(k)]
        gamma_array = np.array([mn_tmp[i].pdf(X)*pi_array[i] for i in range(k)]).T
        gamma_array /= gamma_array.sum(axis = -1).reshape(-1,1)
        
        ## update pi, mu, Sigma ##
        pi_array = gamma_array.sum(axis = 0) / N
        mu_array = gamma_array.T.dot(X) / gamma_array.sum(axis = 0).reshape(-1,1)
        sigma_tmp = []
        
        for i in range(k):
            tmp = X - mu_array[i]  
            tmp = tmp.T.dot(tmp * gamma_array[:,i].reshape(-1, 1)) /( gamma_array.sum(axis = 0)[i]) 
            sigma_tmp.append(tmp)
        
        sigma_array = np.array(sigma_tmp)

        
    return (pi_init, mu_init, sigma_init), (pi_array, mu_array, sigma_array)

def get_gamma(X, para_final):
    pi_final, mu_final, sigma_final = para_final
    k = len(mu_final)
    mn_tmp = [mn(mu_final[i], sigma_final[i]) for i in range(k)]
    gamma_array = np.array([mn_tmp[i].pdf(X) * pi_final[i] for i in range(k)])
    
    return gamma_array

def assign_by_para(X, para_final):
    gamma_array = get_gamma(X, para_final).T
    error = gamma_array.max(axis = -1).sum()
    gamma_array = (gamma_array == gamma_array.max(axis = -1).reshape(-1,1))
    C_list = [X[gamma_array[:, i], :] for i in range(len(gamma_array.T))]
    return C_list, error

def plot_contour(C, para, ax):
    bounds = np.concatenate((C.min(axis = 0).reshape(-1, 1), C.max(axis = 0).reshape(-1, 1)), axis = -1)
    xmin, xmax = bounds[0][0], bounds[0][1]
    ymin, ymax = bounds[1][0], bounds[1][1]    
    x = np.arange(xmin, xmax, 0.01)
    y = np.arange(ymin, ymax, 0.01)    
    x, y = np.meshgrid(x, y)
    xy = np.concatenate((x.reshape(x.shape + (1,)), y.reshape(y.shape + (1,))), axis = -1)
    z = get_gamma(xy, para).sum(axis = 0)
    levels = MaxNLocator(nbins = 10).tick_values(z.min(), z.max())
    ax.contour(x, y, z, alpha = 0.6, levels = levels, cmap = 'jet')
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)


def plot(C_list, para_init, para_final):
    fig, axes = plt.subplots(1, 2, figsize = (20, 8))
    C = np.concatenate((C_1, C_2, C_3), axis = 0)
    pi_init, mu_init, sigma_init = para_init
    pi_final, mu_final, sigma_final = para_final
      
    plot_contour(C, para_init, axes[0])    
    for c in C_list:
        axes[0].scatter(c.T[0], c.T[1] ,alpha = 0.5)
    axes[0].scatter(mu_init.T[0], mu_init.T[1], marker = "x", s = 300, linewidth = 5, color = "r")
    axes[0].set_title("Correct clustering", fontsize = 15)
    
    plot_contour(C, para_final, axes[1])        
    C_list_new, _ = assign_by_para(C, para_final)     
    for c in C_list_new:
        axes[1].scatter(c.T[0], c.T[1] ,alpha = 0.5)
    axes[1].scatter(mu_final.T[0], mu_final.T[1], marker = "x", s = 300, linewidth = 5, color = 'r')
    axes[1].set_title("Fitting clustering", fontsize = 15)


```

```python
NUM_C_1 = 500
NUM_C_2 = 500
NUM_C_3 = 500

C_1 = np.random.multivariate_normal([-2,2], [[3,1],[1,1]] , NUM_C_1)
C_2 = np.random.multivariate_normal([6,1], [[3,1],[1,3]] , NUM_C_2)
C_3 = np.random.multivariate_normal([2,-2], [[1,0],[0,1]] , NUM_C_3)
C_list = [C_1, C_2, C_3]
C = np.concatenate((C_1, C_2, C_3), axis = 0)

para_init, para_final = EM(C, 3, 100, is_random = False, preset_centers = [[-2,0], [-3,0], [-4,0]])
plot(C_list, para_init, para_final)

para_init, para_final = EM(C, 3, 100, is_random = False, preset_centers = [[8,-2], [1,-4], [-5, -1]])
plot(C_list, para_init, para_final)
```

## K-means
### Algebra
As aformentioned, K-means is a special case of Guassian mixture model. Parameters are only $\boldsymbol{\mu}_1, ..., \boldsymbol{\mu}_K$ in K-means. For K-means, all $\Sigma_k$ is set to be close to zero ($\epsilon$). With such infinitely small variance, we will have:

$$
\gamma(z_{nj}) = \frac{\pi_j\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_j, \epsilon)}{\sum_{k}^K\pi_k\mathscr{N}(\boldsymbol{x}_n|\boldsymbol{\mu}_k, \epsilon)}
= \frac{\pi_j\exp(- ||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2/2\epsilon)}{\sum_{k}^K\pi_k\exp(- ||\boldsymbol{x}_n - \boldsymbol{\mu}_k||^2/2\epsilon)}
$$

In the denominator, only the smallest $||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2$ will go to zero most slowly and finally survive. All in orther words,

$$
\gamma(z_{nj}) = 1, if\ ||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2\ is\  the\  smallest,\ otherwise\ 0
$$

This is exactly the **E-step** of K-means, then in term of $\mathbb{E}_Z(\boldsymbol{\theta})$,

$$
\mathbb{E}_Z(\boldsymbol{\theta}) = \sum_{n=1}^N\sum_{j=1}^K\gamma(z_{nj})(\ln \pi_j - ||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2/2\epsilon)
$$

where $||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2/2\epsilon$ will be far bigger than $\ln \pi_j$, then

$$
\mathbb{E}_Z(\boldsymbol{\theta}) = -\sum_{n=1}^N\sum_{j=1}^K\frac{1}{2}\gamma(z_{nj})||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2
$$

Note that $\epsilon$ can be ignored as a constant, and this is the **M-step** of K-means.

To make it more concise:
1. Initialize $\boldsymbol{\mu}_k$
2. Assign $\gamma(z_{nj})$, usually call $r_{nj}$ in K-means; $r_{nj} = \mathbb{1}\{j=\mathop {argmin}_{k}\ ||\boldsymbol{x}_n - \boldsymbol{\mu}_k||^2\}$
3. Assign $\boldsymbol{\mu}_k = \frac{\sum_{n=1}^Tr_{nk}\boldsymbol{x}_n}{\sum_{n=1}^Tr_{nk}}$
4. Repeat 2, 3 until $-\sum_{n=1}^N\sum_{j=1}^K\frac{1}{2}\gamma(z_{nj})||\boldsymbol{x}_n - \boldsymbol{\mu}_j||^2$ converges

Since K-means can be regarded as a special case of Guassian mixture model, the underlying assumption is each cluster obeys Normal distribution

### Simulation

In the simulation, two parameters are tuned, i.e. initialization of $\boldsymbol{\mu}$ and cluster number k. As aformentioned, EM algorithm can only find the local optimization, thus intialization will affect the final results. With increasing k, the value of loss function will decrease in an "elbow" way and the k value corresponding to the turnning point is usually the best choice.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mn


def initialize_mu(X, k):
    dim = X.shape[-1]
    bounds = np.concatenate((X.min(axis = 0).reshape(-1, 1), X.max(axis = 0).reshape(-1, 1)), axis = -1)
    mu = []
    for d in range(dim):
        mu.append(np.random.uniform(bounds[d][0], bounds[d][1], k))
    mu_array = np.array(mu).T
    return mu_array

def EM(X, k, iteration = 100, is_random = True, preset_centers = None):
    ## Initialize mu ##
    dim = X.shape[-1]
    N = len(X)
    if is_random:
        mu_array = initialize_mu(X, k)      
    else:
        mu_array = np.array(preset_centers)
    
    mu_init = mu_array.copy()
         
    for _ in range(iteration):
        ## assign gamma ##
        gamma_array = np.array([np.linalg.norm(X - mu_array[i], axis = -1) for i in range(k)]).T
        gamma_array = (gamma_array == gamma_array.min(axis = -1).reshape(-1,1))

        ## update mu ##
        mu_array = gamma_array.T.dot(X) / gamma_array.sum(axis = 0).reshape(-1,1)
        
    return mu_init, mu_array

def assign_by_mu(X, mu_final):
    K = len(mu_final)
    gamma_array = np.array([np.linalg.norm(X - mu_final[i], axis = -1) for i in range(K)]).T
    error = gamma_array.min(axis = -1).sum()
    gamma_array = (gamma_array == gamma_array.min(axis = -1).reshape(-1,1))
    
    C = [X[gamma_array[:, i], :] for i in range(K)]
    return C, error
        
def plot(C_list, mu_init, mu_final):
    fig, axes = plt.subplots(1, 2, figsize = (20, 8))
    for c in C_list:
        axes[0].scatter(c.T[0], c.T[1] ,alpha = 0.5)
    axes[0].scatter(mu_init.T[0], mu_init.T[1], marker = "x", s = 300, linewidth = 5)
    axes[0].set_title("Correct clustering", fontsize = 15)
    
    C = np.concatenate((C_1, C_2, C_3), axis = 0)
    C, _ = assign_by_mu(C, mu_final)
    for c in C:
        axes[1].scatter(c.T[0], c.T[1] ,alpha = 0.5)
    axes[1].scatter(mu_final.T[0], mu_final.T[1], marker = "x", s = 300, linewidth = 5)
    axes[1].set_title("Fitting clustering", fontsize = 15)
    
    
    
```

```python
NUM_C_1 = 500
NUM_C_2 = 500
NUM_C_3 = 500

C_1 = np.random.multivariate_normal([-2,2], [[3,1],[1,1]] , NUM_C_1)
C_2 = np.random.multivariate_normal([6,1], [[3,1],[1,3]] , NUM_C_2)
C_3 = np.random.multivariate_normal([2,-2], [[1,0],[0,1]] , NUM_C_3)
C_list = [C_1, C_2, C_3]

C = np.concatenate(C_list, axis = 0)
i, f = EM(C, 3, 1000, False, [[-2,0], [-3,0], [-4,0]])
plot(C_list, i, f)
i, f = EM(C, 3, 1000, False, [[8,-2], [1,-4], [-5, -1]])
plot(C_list, i, f)

fig, axes = plt.subplots(2, 3, figsize = (20, 10))
fig.suptitle('Fitting with different number of clusters K', fontsize = 30)
axes = axes.reshape(-1)
errors = []
for k in range(1, 6):
    i, f = EM(C, k, 1000, True)
    C_asigned, e = assign_by_mu(C, f)
    errors.append(e)
    for c in C_asigned:
        axes[k-1].scatter(c.T[0], c.T[1] ,alpha = 0.3)
    axes[k-1].scatter(f.T[0], f.T[1], marker = "x", s = 300, linewidth = 5, color = "r")
    
axes[5].plot(np.arange(1, 6), errors)
```
